```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SIZE = 200

image = np.zeros((SIZE,SIZE, 3), dtype=int)
for i in range(SIZE):
    for j in range(SIZE):
        red = (i * j) % 256
        green = (i * i -j*j) % 256
        blue = (i*i + j*j) % 256
        image[i,j] = [red,blue,green]
        # image[i,j] = (i*j) % 201

plt.figure(figsize=(10,10))
plt.imshow(image, cmap = 'Spectral', interpolation='hamming')
plt.show()
```
